# Example 1: Simple regression model

```python
from tensorflow import keras
import matplotlib.pyplot as plt

X = [-2,-1,0,1,2,3]

Y = [-8,-5,-2,1,4,7]

plt.plot(X,Y)
```

```python
def f(x):
  return 3*x - 2;

y = f(x = 10)
print(f"y = {y}")
```

## Simplest Possible Neural Network

```python
model = keras.Sequential([

    keras.layers.Dense(units = 1, input_shape = [1])


])


model.summary()
```

```python
model.compile(

    optimizer = keras.optimizers.SGD(learning_rate=0.01),
    loss = "mean_squared_error"

)
```

# Before Training

```python
print(f"{model.predict([10.0])}")
```

```python
w, b = model.get_weights()
print(f'Before Training... \n weight = {w} \n bias = {b}')
```

```python
10.0 * w + b
```

```python
hist = model.fit(X,Y, epochs = 750)
```

```python
plt.title('Training ' + 'loss')
plt.plot(hist.history['loss'])
plt.show()
```

```python
y = f(x = 10)
print(f'Expected Output = {y}')
```

```python
pred = model.predict([10.0])

print(f"Actual Output = {pred[0][0]}")
```
